### Text mining analysis. Let's check the relative frequency of keywords assigned to the four broad categories:
* insight
* health
* community
* individual

```python
insight='''insight
meditation
feel
mind
mindfulness
think
know
thoughts
practice
experience
feeling
focus
feelings
present
experiences
study
mindful
attention
research
emotions
change'''.split('\n')
insight = list(set(insight))

health='''health
body
wellbeing
life
help
anxiety
brain
body
sleep
stress
exercise
results
positive
physical
suffering
mental'''.split('\n')
health = list(set(health))

community='''community
people
work 
world
social
love
relationship
friends
home
family'''.split('\n')
community = list(set(community))

individual='''individual
personal
self
myself'''.split('\n')
individual = list(set(individual))

time='''time
day
years
moment
days
long
minutes
times'''.split('\n')
time = list(set(time))
```

```python
cats=[insight,health,community,individual,time]
li=[]

for cat in cats:
    li.append(freq[freq['word'].isin(cat)])
```

```python
cats_str=['insight','health','community','individual','time']

for no, cat in enumerate(cats_str):
    li[no]['cat']=cat
    
df=pd.DataFrame(columns=['word','count','freq','cat'])
df=df.append(li)
```

```python
df['count']=df['count'].astype(int)
df['freq']=df['freq'].round(1)

dfGr=df.groupby('cat').sum()
dfGr['keywords']=df.groupby('cat').size()
dfGr['count_per_keyword']=dfGr['count']/dfGr['keywords']
dfGr['count_per_keyword']=dfGr['count_per_keyword'].round(1)
```
